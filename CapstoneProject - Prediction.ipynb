```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns 

from sklearn.preprocessing import StandardScaler

from sklearn import preprocessing, metrics, tree, svm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.externals.six import StringIO
from sklearn.metrics import classification_report, confusion_matrix, jaccard_similarity_score, log_loss, f1_score
from sklearn.linear_model import LogisticRegression,LinearRegression 

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Import Dataframe
df = pd.read_csv('df_FeatureSelection.csv')
```

```python
df.drop('SalePrice', axis = 1, inplace = True)
df.head()
```

```python
# Create X and y dataframes
X = df.iloc[:,1:]
X.drop('SalePrice_log', axis = 1, inplace = True)
y = df['SalePrice_log']

# Drop Neighborhood columns
X = X[X.columns.drop(list(X.filter(regex='Neighborhood')))]

# STANDARDIZE
X = preprocessing.StandardScaler().fit(X).transform(X)

# Split into test and train set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

#### Linear Regression

```python
reg_liner=LinearRegression()
reg_liner.fit(X_train, y_train)
reg_liner.score(X_train, y_train)
```

#### Logistic regression

```python
reg_logistic = LogisticRegression()
reg_logistic.fit(X_train, y_train)
print(reg_logistic.score(X_train, y_train))
```
